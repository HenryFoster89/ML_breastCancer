```python
import pandas as pd
import matplotlib.pyplot as plt

from utils.config import cols_name

df = pd.read_csv(r"data\wdbc.data", index_col=False, names = cols_name)

df.head(10).to_clipboard()
```

# Enrico ToDo

```python


# Conta i valori unici nella colonna 'diagnosis'
diagnosis_counts = df["diagnosis"].value_counts()

# Creazione del bar plot
plt.figure(figsize=(6, 4))
plt.bar(diagnosis_counts.index, diagnosis_counts.values, color=['blue', 'orange'])
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.title("Count of Unique Values in 'Diagnosis' Column")
plt.show()
```
